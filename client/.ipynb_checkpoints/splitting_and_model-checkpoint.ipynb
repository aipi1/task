# Obtaining full data

```python
import numpy as np
import pandas as pd
# from datetime import datetime
```

```python
data = pd.read_csv('../task/ds1.csv', parse_dates = ['date'])
data
```

```python
data.isnull().sum().any()
```

```python
data.sort_values(by = 'date', inplace = True, ascending = True)
data
```

```python
# t = datetime.now()

# print("now =", t)
```

```python
data.set_index('date', inplace = True)
data.head()
```

```python
y = data.y.copy()
X = data.drop('y', axis = 1).copy()
X.head()
```

```python
y.head()
```

# Splitting data to test and train sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle = False)
X_train
```

```python
X_test
```

```python
X_train.to_csv('X_train.csv', index = True)
X_test.to_csv('X_test.csv', index = True)
y_train.to_csv('y_train.csv', index = True)
y_test.to_csv('y_test.csv', index = True)
```

# Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error

mae = []
tscv = TimeSeriesSplit(n_splits = 5)
for train_index, test_index in tscv.split(X, y):
    model = LinearRegression()
    model.fit(X.iloc[train_index], y.iloc[train_index])
    preds = model.predict(X.iloc[test_index])
    mae.append(mean_squared_error(y.iloc[test_index], preds))

print(np.mean(mae))
```

```python
import pickle
pickle.dump(model, open('../server/model.pickle', 'wb'))
```
